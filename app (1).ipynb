#### 3. Load (BookDWH)

```python
dwh_conn = pyodbc.connect('Driver={SQL Server};'
    'Server=localhost;'
    'Database=BookDWH;'
    'Trusted_Connection=yes;')

cursor = dwh_conn.cursor()    
```

```python
df_customer.dtypes
```

```python
for index, row in df_customer.iterrows():
    cursor.execute('INSERT INTO DimCustomer VALUES(?, ?, ?)', row['CustomerID'], row['FullName'], row['City'])
dwh_conn.commit()
```

```python
df_book.head(5)
```

```python
for index, row in df_book.iterrows():
    cursor.execute('INSERT INTO DimBook VALUES(?, ?, ?, ? ,? )', row['BookID'], row['Title'], row['Price'], row['CategoryDescription'], row['AuthorName'])
dwh_conn.commit()
```

```python
df_factorder.head(5)
```

```python
df_dimbook = pd.read_sql ("select * from DimBook", dwh_conn)
df_dimbook.head(5) 
```

```python
df_factorder = pd.merge(df_factorder, df_dimbook, how='inner', on='BookID')
```

```python
df_factorder = df_factorder[["OrderID", "CustomerID", "OrderDate", "BookSK", "TotalAmount", "Quantity"]] 
```

```python
df_factorder.head(5)
```

```python
for index, row in df_factorder.iterrows():
    cursor.execute('INSERT INTO FactOrders VALUES(?, ?, ?, ? ,?, ? )', row['OrderID'], row['CustomerID'], row['OrderDate'], row['BookSK'], row['TotalAmount'], row['Quantity'])
dwh_conn.commit()
```
